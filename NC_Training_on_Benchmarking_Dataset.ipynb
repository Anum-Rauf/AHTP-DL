## As we have four types of g-gap DPC (with gap 0,1,2, and 3) features, from these four DPC features we  have created four different combinations and created RGB images from them. in next 4 steps four types of images are created. 

````
#Images for DPC features with gap 0,1,2
a = np.reshape(dataaa0,[1826,20,20])
b = np.reshape(dataaa1,[1826,20,20])
c = np.reshape(dataaa2,[1826,20,20])
feature1=np.zeros((1826,20,20,3))
for i in range(1826):
  data = np.zeros( (20,20,3))
  data[:,:,0]=a[i,:,:]
  data[:,:,1]=b[i,:,:]
  data[:,:,2]=c[i,:,:]
  feature1[i]=data     
````

````
#Images for DPC features with gap 0,1,3
a = np.reshape(dataaa0,[1826,20,20])
b = np.reshape(dataaa1,[1826,20,20])
c = np.reshape(dataaa3,[1826,20,20])
feature2=np.zeros((1826,20,20,3))
for i in range(1826):
  data = np.zeros( (20,20,3))
  data[:,:,0]=a[i,:,:]
  data[:,:,1]=b[i,:,:]
  data[:,:,2]=c[i,:,:]
  feature2[i]=data 
````

````
#Images for DPC features with gap 0,2,3
a = np.reshape(dataaa0,[1826,20,20])
b = np.reshape(dataaa2,[1826,20,20])
c = np.reshape(dataaa3,[1826,20,20])
feature3=np.zeros((1826,20,20,3))
for i in range(1826):
  data = np.zeros( (20,20,3))
  data[:,:,0]=a[i,:,:]
  data[:,:,1]=b[i,:,:]
  data[:,:,2]=c[i,:,:]
  feature3[i]=data 
````

````
#Images for DPC features with gap 1,2,3
a = np.reshape(dataaa1,[1826,20,20])
b = np.reshape(dataaa2,[1826,20,20])
c = np.reshape(dataaa3,[1826,20,20])
feature4=np.zeros((1826,20,20,3))
for i in range(1826):
  data = np.zeros( (20,20,3))
  data[:,:,0]=a[i,:,:]
  data[:,:,1]=b[i,:,:]
  data[:,:,2]=c[i,:,:]
  feature4[i]=data
````

# Generating four CNN models and saving them in drive

````
#%% Training CNN for First type of Images (Gap with 0,1,2 (Features1))
seed = 1
numpy.random.seed(seed)
kfold = StratifiedKFold(n_splits=10, shuffle=True, random_state=seed)
cvscores = []
testing=[]
pred=[]
i=1
for train, test in kfold.split(features1, labels): #.......................................................
  model = Sequential()
  model.add(Conv2D(32, (3, 3),  input_shape=features1[0].shape))#.......................................................
  convout1 = Activation('relu')
  model.add(convout1)
  convout3 = Activation('relu')
  model.add(convout3)
  model.add(Conv2D(8,( 3, 3)))
  
  model.add(MaxPooling2D(pool_size=(2,2 )))
  model.add(Dropout(0.5))

  model.add(Flatten())
  model.add(Dense(170))
  model.add(Activation('relu'))
  model.add(Dropout(0.5))
  model.add(Dense(1))
  model.add(Activation('sigmoid'))
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=["accuracy"])
  ii=str(i)
  filepath="/content/drive/My Drive/Colab Notebooks/new-weights-bench-cv10-0,1,2-NC-"+ii+".hdf5"  #.......................................................
  checkpoint = ModelCheckpoint(filepath, monitor='val_accuracy', verbose=1, save_best_only=True, mode='max', period=1)
  callbacks_list = [checkpoint]
	# Fit the model
  model.fit(features1[train], labels[train], batch_size=35, epochs=370, verbose=1, validation_data=(features1[test], labels[test]), callbacks=[checkpoint]) #.......................................................
  score2=model.evaluate(features1[train],labels[train],verbose=0) #..............................
  print("acuuurcy:")
  print(score2)
  model = Sequential()
  model.add(Conv2D(32, (3, 3),  input_shape=features1[0].shape)) #.......................................................
  convout1 = Activation('relu')
  model.add(convout1)
  convout3 = Activation('relu')
  model.add(convout3)
  model.add(Conv2D(8,( 3, 3)))
  
  model.add(MaxPooling2D(pool_size=(2,2 )))
  model.add(Dropout(0.5))
  model.add(Flatten())
  model.add(Dense(170))
  model.add(Activation('relu'))
  model.add(Dropout(0.5))
  model.add(Dense(1))
  model.add(Activation('sigmoid'))
  #model.compile(loss='binary_crossentropy', optimizer='adam', metrics=["accuracy"])
# load weights
  model.load_weights(filepath)
# Compile model (required to make predictions)
  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=["accuracy"])
  print("Created model and loaded weights from file")
# load pima indians dataset
  score2=model.evaluate(features1[train],labels[train],verbose=0) #.......................................................
  print("acuuuraaacy:")
  print(score2)
  scores = model.evaluate(features1[test], labels[test], verbose=0) #.......................................................
  print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
  cvscores.append(scores[1] * 100)
  p = model.predict_classes(features1[test])  #.......................................................     
  pred.append(p)
  testing.append(labels[test]) 
  i=i+1
print("%.2f%% (+/- %.2f%%)" % (numpy.mean(cvscores), numpy.std(cvscores)))
````

````
# Calculating performance estimators for first CNN model
from sklearn.metrics import matthews_corrcoef
#import keras_metrics
#from keras import backend as K
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
#import keras_metrics as km
from sklearn.metrics import f1_score 
a = np.array(pred)
aa= []
for sublist in a:
    for val in sublist:
        aa.append(val)

b= np.array(testing)
bb= []
for sublist in b:
    for val in sublist:
        bb.append(val)
sensitivity1=np.zeros(10) #...........................
specificity1=np.zeros(10) #...........................
mcc=np.zeros(10)  #...........................
accuracy1=np.zeros(10) #...........................
aucc=np.zeros(10) #...........................
tp=np.zeros(10) #...........................
fp=np.zeros(10) #...........................
fn=np.zeros(10) #...........................
tn=np.zeros(10) #...........................
for i in range(10): #...........................
  print('All observations for model', i)
  #%% sensitivity and specificity
  cm1 = confusion_matrix(b[i], a[i])
  sensitivity1[i] = cm1[0,0]/(cm1[0,0]+cm1[0,1])
  print('Sensitivity : ', sensitivity1[i] )

  specificity1[i] = cm1[1,1]/(cm1[1,0]+cm1[1,1])
  print('Specificity : ', specificity1[i])
  tp[i]=cm1[0,0]
  print('tp : ', tp[i])
  fp[i]=cm1[0,1]
  print('fp : ', fp[i])
  fn[i]=cm1[1,0]
  print('fn : ', fn[i])
  tn[i]=cm1[1,0]
  print('tn : ', tn[i])
  #%%
  from sklearn.metrics import matthews_corrcoef
  mcc[i] = matthews_corrcoef(b[i],a[i])
  print('Matthews Correlation Coefficient:', mcc[i])

  #%%
  total1=sum(sum(cm1))

  accuracy1[i]=(cm1[0,0]+cm1[1,1])/total1
  print ('Accuracy : ', accuracy1[i])

  import numpy as np
  from sklearn import metrics
  fpr, tpr, thresholds = metrics.roc_curve(b[i], a[i])
  aucc[i]=metrics.auc(fpr, tpr)
  print('AUC:', aucc[i])
  
print("AVG OF ALL:")
print('Sensitivity : ', np.mean(sensitivity1) )
print('Specificity : ', np.mean(specificity1))
print('Matthews Correlation Coefficient:', np.mean(mcc))
print('AUC:', np.mean(aucc))
print ('Accuracy : ', np.mean(accuracy1))
print('tp:',np.mean(tp))
print('fp:',np.mean(fp))
print('tn:',np.mean(tn))
print('fn:',np.mean(fn))
````
